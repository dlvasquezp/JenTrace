# Getting started
In this simple example, a simple optical design composed by an achromat is created to trace a point source to the focal plane. At the end, the spot diagram and Seidel aberration is calculated and displayed. 

## Optical system
To create an optical system, the 'OpSysData' class is imported. Once instantiated, it will create two default surfaces i.e. Object and Image surfaces. From this point, the information of each optical surface is provided in the following way:

    (Thickness, Curvature (or 1/Radius), Refraction index (or material name))

For practical purposes, an achromat from the manufacturer Edmund Optics has been selected and is going to be used to trace the rays. Link https://www.edmundoptics.de/p/20mm-dia-x-35mm-fl-mgfsub2sub-coated-achromatic-doublet-lens/5821/

Note that the surface diameter information is defined as radius in the variable 'clearSemDia'. This values have only a cosmetic purpose (it has no influence in the calculations) and will allow us to better undertand the system. If the variable 'clearSemDia' fail to match the number of surfaces, a default vaulue will be used to display the system.

```python
from opt_sys import OpSysData
syst1 = OpSysData()

syst1.add_surface(7.33 ,1/24.24   ,'S-BAH11')
syst1.add_surface(1.76 ,1/-15.96  ,'N-SF10')
syst1.add_surface(5    ,1/-139.78 ,1)

#Modify object distance (surfIndex #0)
syst1.change_surface(45.5 ,0 ,1 ,surfIndex=0)

#Show system
clearSemDia=[20/2,20/2,20/2,20/2,20/2]
syst1.print_report()
syst1.plot_optical_system(clearSemDia)
```

## Point Source
In the pyTrace module, a RaySource is defined as list of rays, which are defined in the following way:

    ([X,Y,Z],[CosX,CosY,CosY],wavelength)
    XYZ: Position [mm] with respect to the Object surface 
    CosX/Y/Z: cosine directions

A 'PointSource' is a RaySource subclass, where the rays share same origin (position) and wavelenght. Therefore to instantiate it only the position and wavelenght is required. 

    PointSource([X,Y,Z], wavelength)

Note: When instatiated, it will create automatically five rays, which are directed to the center and cardinal poinst of the aperture stop, see "For Developers" secction. Once the optical design is solved, the ray directoins are calculated.

```python
from ray_src import PointSource
#PointSource: Ray source with a common starting point.
pto1=PointSource([0,4,0],587.6)

#Show rays
pto1.print_report()
```

## Optical design
An optical design consist on three main parts: rays information (Point Source),  surface information (Optical system) and Aperture stop (radius and surface index). 

    OpticalDesign(PointSource,OpticalSystem,ApertureRadius,ApertureSurfaceIndex)

With this data, the optical design can propagate the rays and iterate using optimization functions to find the information of principal ray, marginal rays, coma rays, etc. When the characteristic parameters of the system are found, the optical design is meant to be solved.

```python
from opt_dsg import OpDesign
#Aperture radius
aprRad = 8.0 #mm
#Aperture position (Surface index)
aprInd = 3
#Create an optical design. 
design1  = OpDesign(pto1,syst1,aprRad,aprInd)

#Plot design
design1.usrSrc.print_report()
design1.optSys.print_report()
design1.plot_design(clearSemDia)
```

## Autofocus
Once the optical design is solved, the built-in function autofocus can be used to find the image position where the PointSource has the minimun mean square-root radius

```python
design1.autofocus() 

#Plot design
design1.usrSrc.print_report()
design1.optSys.print_report()
design1.plot_design(clearSemDia)
```

## Spot diagram
To create the spot diagram, a pseudo-random algorithm is used to create rays inside the aperture stop area. Later on, the rays are propagated and the desired sport diagram is generated.

```python
from spt_dgm import spot_diagram
#Create a spot diagram with pseudo random initial distribution
spotSrc,spotTrace = spot_diagram(design1,noRays=1000,show=True,surfIndex=-1)
```

## Seidel Coefficients
Since the Optical design is solved, the Seidel coefficents can be calculated using the chief and marginal rays.

```python
from abr_fnc import seidel_coef, plot_seidel
#Calculate seidel coefficeints for primary aberrations
SCoef =  seidel_coef(design1)
plot_seidel(SCoef)
```
